```python
import pandas as pd

df = pd.read_csv('e468_midterm_20.txt', sep='  ')
```

```python
from statsmodels.formula.api import ols
import numpy as np
```

```python
fit1 = ols('C ~ Y', data=df).fit()
fit1.summary()
```

```python
s1,s2,s3,s4 = np.full(100, 0), np.full(100, 0), np.full(100, 0), np.full(100, 0)

i=0
while i<100:
    s1[i], s2[i+1], s3[i+2], s4[i+3] = 1, 1, 1, 1
    i += 4
    
df['s1'], df['s2'], df['s3'], df['s4'] = s1, s2, s3, s4
```

```python
fit2 = ols('C ~ s1 + s2 + s3 + s4 + Y -1', data=df).fit()
fit2.summary()
```

```python
fit3 = ols('C ~ s1 + s2 + s3 + s4 -1', data=df).fit()
df['residuals_consumption'] = fit3.resid
fit4 = ols('Y ~ s1 + s2 + s3 + s4 -1', data=df).fit()
df['residuals_income'] = fit4.resid
```

```python
fit5 = ols('residuals_consumption ~ residuals_income -1', data=df).fit()
fit5.summary()
```

```python
df['sn1'], df['sn2'], df['sn3'], = df.s1 - df.s4, df.s2 - df.s4, df.s3 - df.s4
```

```python
fit6 = ols('C ~ sn1 + sn2 + sn3 + Y', data=df).fit()
fit6.summary()
```
